## ***1. Know Your Data***

```python
import pandas as pd
import numpy as np
import sqlite3
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load data from SQLite
conn = sqlite3.connect('../phonepe_pulse.db')
agg_trans = pd.read_sql("SELECT * FROM aggregated_transaction", conn)
agg_user = pd.read_sql("SELECT * FROM aggregated_user", conn)
map_trans = pd.read_sql("SELECT * FROM map_transaction", conn)
map_user = pd.read_sql("SELECT * FROM map_user", conn)
top_trans = pd.read_sql("SELECT * FROM top_transaction", conn)
top_user = pd.read_sql("SELECT * FROM top_user", conn)
conn.close()
print("Data loaded successfully!")
```

## ***2. Data Analysis & Storytelling***

### **Chart 1: Year-wise Total Transaction Amount**

```python
df1 = agg_trans.groupby('Year')['Transaction_Amount'].sum().reset_index()
fig1 = px.bar(df1, x='Year', y='Transaction_Amount', title='Year-wise Total Transaction Amount', color='Transaction_Amount')
fig1.show()
```

### **Chart 2: State-wise Transaction Distribution**

```python
df2 = agg_trans.groupby('State')['Transaction_Amount'].sum().sort_values(ascending=False).reset_index().head(10)
fig2 = px.pie(df2, names='State', values='Transaction_Amount', title='Top 10 States by Transaction Amount')
fig2.update_traces(textinfo='percent+label')
fig2.show()
```

### **Chart 3: Most Popular Transaction Types**

```python
df3 = agg_trans.groupby('Transaction_Type')['Transaction_Count'].sum().reset_index()
fig3 = px.funnel(df3, x='Transaction_Count', y='Transaction_Type', title='Transaction Count by Type')
fig3.show()
```

### **Chart 4: User Growth Trends Across Regions**

```python
df4 = map_user.groupby(['Year', 'Quarter'])['RegisteredUsers'].sum().reset_index()
df4['Time'] = df4['Year'].astype(str) + " Q" + df4['Quarter'].astype(str)
fig4 = px.line(df4, x='Time', y='RegisteredUsers', title='Registered User Growth Over Time (Quarterly)', markers=True)
fig4.show()
```

### **Chart 5: Top Mobile Brands Used by Customers**

```python
df5 = agg_user.groupby('Brand')['Count'].sum().sort_values(ascending=False).reset_index().head(10)
fig5 = px.treemap(df5, path=['Brand'], values='Count', title='Top 10 Mobile Brands by User Base')
fig5.show()
```

### **Chart 6: Transaction Volume vs Amount Heatmap**

```python
# Correlation analysis
corr = agg_trans[['Transaction_Count', 'Transaction_Amount']].corr()
fig6 = px.imshow(corr, text_auto=True, aspect="auto", title="Correlation: Count vs Amount")
fig6.show()
```

### **Conclusion**

The EDA reveals a massive surge in India's digital payments, with transaction amounts showing consistent upward trends. States with higher urbanization tend to dominate the transaction volume, while Peer-to-Peer and Merchant payments remain the most popular categories.
